```python
# Atv 7 - A e B
from matplotlib import pyplot as plt

dataset = [10, 15, 20, 25, 30, 35, 40]
frequencias = [3, 5, 8, 6, 4, 2, 1]

plt.title('Gráfico de Barras')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.bar(dataset, frequencias)
plt.show()

plt.title('Gráfico de Pizza')

sum = 0
for i in range(len(dataset)):
  sum += frequencias[i]

sub_dataset = []
for i in range(len(dataset)):
  sub_dataset.append(round((frequencias[i]/sum)*100))

plt.pie(frequencias, labels=sub_dataset)
plt.show()
```

```python
# Atv 12 - A, B e C
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt

idades = np.random.randint(18, 70, size=50)
idades2 = np.random.randint(18, 70, size=10000)

# A
print(f"Média = {mean(idades)}")
print(f"Mediana = {median(idades)}")
print(f"Moda = {stats.mode(idades)}")

# B
plt.hist(idades, bins=10)
plt.show()

plt.boxplot(idades)
plt.show()

# C
print(f"Média = {mean(idades2)}")
print(f"Mediana = {median(idades2)}")
print(f"Moda = {stats.mode(idades2)}")

plt.hist(idades2, bins=10)
plt.show()

plt.boxplot(idades2)
plt.show()
```
